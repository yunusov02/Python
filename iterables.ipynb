```python
colors = ['red', 'green', 'blue']
color_iters = iter(colors)

color = next(color_iters)
print(color)
```

```python
bonuses = [100, 200, 300]
new_bonuses = []

for bonus in bonuses:
    new_bonuses.append(bonus*2)

print(new_bonuses)
```

```python
def double_power(x):
    return x ** 2

bonuses = [100, 200, 300]
iterator = map(double_power, bonuses)

print(next(iterator))
print(next(iterator))
print(next(iterator))
```

```python
# anotor example using map function
names = ['david', 'peter', 'john']

iterator = map(lambda name: name.capitalize(), names)

print(next(iterator))
print(next(iterator))
print(next(iterator))
```

```python
scores = [34, 90, 50, 78, 80]

filtered_scores = filter(lambda x: x >= 70, scores)

print(next(filtered_scores))
print(next(filtered_scores))
print(next(filtered_scores))
```

```python
from functools import reduce

def sum(a, b):
    print(f"a={a}, b={b}, {a} + {b} = {a + b}")
    return a + b

scores = [75, 65, 80, 95, 50]

total = reduce(sum, scores)
print(total)
```

```python
from functools import reduce
scores = [75, 80, 75, 90]

total = reduce(lambda a, b: a + b, scores)
print(total)
```

<h1>List Comprehension</h1>

```python
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x ** 2, numbers))
print(squares)
```

```python
numbers = [1, 2, 3, 4, 5]
squares = [number ** 2 for number in numbers]
print(squares)
```

```python
mountains = [
    ['Makalu', 8485],
    ['Lhotse', 8516],
    ['Kanchendzonga', 8586],
    ['K2', 8611],
    ['Everest', 8848]
]

highest_mountains = list(filter(lambda x: x[1] > 8600, mountains))
highest_mountains
```

```python
highest_mountains = [m for m in mountains if m[1] > 8600]
print(highest_mountains)
```
